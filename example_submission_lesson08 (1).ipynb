```python
# import usual things & new packages we've been using:
import ipywidgets
import numpy as np
import matplotlib.pyplot as plt
import bqplot.pyplot as bplt
import pandas as pd
```

```python
planets = pd.read_csv('https://jnaiman.github.io/csci-p-14110_su2020/lesson08/planets_2020.06.22_10.10.17.csv', 
                     sep=",", comment="#")
```

```python
planets
```

```python
#plot planet density
fig, ax=plt.subplots(1,1,figsize=(4,4))
ax.hist(planets['pl_dens'])
ax.set_xlabel('Planet Density [g/cm**3]')
ax.set_ylabel('Frequency')
ax.set_yscale('log')
plt.show()
```

```python
#plot planet radius
fig, ax=plt.subplots(1,1,figsize=(4,4))
ax.hist(planets['pl_radj'])
ax.set_xlabel('Planet Radius [Jupiter radii]')
ax.set_ylabel('Frequency')
ax.set_yscale('log')
plt.show()
```

```python
#scatterplot
fig=bplt.figure()
scatter_plot=bplt.scatter(planets['pl_dens'],planets['pl_radj'])
fig
```

```python
#downsample
nsamples=100
randomInds=np.random.choice(range(len(planets)-1), nsamples, replace=False)
planetsDS=planets.iloc[randomInds]
```

```python
planetsDS
```

```python
fig=bplt.figure()
scatter_plot=bplt.scatter(planetsDS['pl_dens'],planetsDS['pl_radj'],interactions={'click':'select'},selected_style={'fill':'purple'})
bplt.xlim(0,15)
bplt.show()
```

```python
planetsDS.iloc[scatter_plot.selected]
```

```python
planetsDS.iloc[scatter_plot.selected]['pl_name'].values
```

```python
planetsDS.iloc[scatter_plot.selected]['pl_name'].values[0]
```

```python
planetName=ipywidgets.Label()
planetName.value=planetsDS.iloc[scatter_plot.selected]['pl_name'].values[0]
```

```python
def on_selection(change):
    print(change)
scatter_plot.observe(on_selection,'selected')
```

```python
fig=bplt.figure()
scatter_plot=bplt.scatter(planetsDS['pl_dens'],planetsDS['pl_radj'],interactions={'click':'select'},selected_style={'fill':'purple'})
bplt.xlim(0,15)
def on_selection(change):
    print(change['new'])
scatter_plot.observe(on_selection,'selected')
fig
```

```python
fig=bplt.figure()
scatter_plot=bplt.scatter(planetsDS['pl_dens'],planetsDS['pl_radj'],interactions={'click':'select'},selected_style={'fill':'purple'})
bplt.xlim(0,15)
def on_selection(change):
    planetName.value=planetsDS.iloc[scatter_plot.selected][change['new']]['pl_name'].values[0]
scatter_plot.observe(on_selection,'selected')
fig
```

```python
fig=bplt.figure()
scatter_plot=bplt.scatter(planetsDS['pl_dens'],planetsDS['pl_radj'],interactions={'click':'select'},selected_style={'fill':'purple'})
bplt.xlim(0,15)
def on_selection(change):
    planetName.value=planetsDS.iloc[change['new']]['pl_name'].values[0]
scatter_plot.observe(on_selection,'selected')
fig.axes[0].label = 'Planet Density' # xaxes label
fig.axes[1].label = 'Planet Radius' # yaxes label
bplt.show()
ipywidgets.VBox([planetName,fig])
```
